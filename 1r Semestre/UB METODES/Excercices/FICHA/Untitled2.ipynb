Because I cannot plot the strings of the different timesdata, i will first convert them the to datetime type:

```python
from datetime import datetime

df['activation_time_local'] = df['activation_time_local'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['last_courier_assigned_time'] = df['last_courier_assigned_time'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['courier_started_order_local'] = df['courier_started_order_local'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['courier_enters_pickup_time_local'] = df['courier_enters_pickup_time_local'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['pickup_time_local'] = df['pickup_time_local'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['courier_enters_delivery_point_time_local'] = df['courier_enters_delivery_point_time_local'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
df['termination_time'] = df['termination_time'].apply(lambda x: datetime.strptime(x, '7/%d/2019 %H:%M:%S'))
```

Now I will forgot about the date, assuming the hole orders are due the same day (I'll deal with exceptions later), and I'm going to transform the datetime type data into an integer representing the seconds into that day:

```python
df['activation_time_local'] = df['activation_time_local'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['last_courier_assigned_time'] = df['last_courier_assigned_time'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['courier_started_order_local'] = df['courier_started_order_local'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['courier_enters_pickup_time_local'] = df['courier_enters_pickup_time_local'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['pickup_time_local'] = df['pickup_time_local'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['courier_enters_delivery_point_time_local'] = df['courier_enters_delivery_point_time_local'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
df['termination_time'] = df['termination_time'].apply(lambda x: x.second + (x.minute * 60) + (x.hour * 3600))
```

Now Im going to plot a histogram of this integers, where we can see that in this case we didn't had a Gaussian, but two Gaussians, one small for launch and the bigger one for dinner time, as one would expect:

    - Most of the orders are placed at dinner with a peak at 77000 seconds into the day more or less, which corresponds to 21:30h (77000/3600 = 21,38 = 21h and 30m).

    - The other relevant peak at launch time, is around 47500 seconds into the day, or 13h (47500/3600 = 13).

```python
(df['activation_time_local']/3600).plot.hist(bins=50, title="Visibility of Activation Time of Orders (hours in the day)")
```

Once we have this done, we can start operating different times and give the rest of the insights you asked, for this, we are going to add the columns:

    - for Delivery Time (courier_enters_delivery_point_time_local - pickup_time_local)
    - for Waiting Time at Pickup (pickup_time_local - courier_enters_pickup_time_local)
    - for Total Time (termination_time - courier_started_order_local)
    - for Max Courier Capacity (60/Total Time)
    
Now it's when I'm going to correct the problems of the orders that took place when 00:00 of midnight happened and the day changed, so the seconds into the day went to 0 from 86400 (24 60 60). To correct this I only need to add this quantity whenever the result is negative:

(Extra: There are some negative values for Waiting Time, that don't come from this change of 00:00, but rather from another source, I assume it came from confusion between Courier Enters Pickup Time and Pickup Time, so I'll just reverse them, if my assumption is wrong, then I would just leave them outside the representation setting the x_range on the histogram plot of [0,40]).

```python
df['Delivery_time'] = df.apply(lambda row: row.courier_enters_delivery_point_time_local - row.pickup_time_local if row.courier_enters_delivery_point_time_local - row.pickup_time_local >= 0 else row.courier_enters_delivery_point_time_local - row.pickup_time_local + 86400 , axis=1)
df['Waiting_time_at_pickup'] = df.apply(lambda row: row.pickup_time_local - row.courier_enters_pickup_time_local if row.pickup_time_local - row.courier_enters_pickup_time_local >= 0 else (row.pickup_time_local - row.courier_enters_pickup_time_local + 86400 if row.pickup_time_local - row.courier_enters_pickup_time_local < -50000 else row.courier_enters_pickup_time_local - row.pickup_time_local) , axis=1)
df['Total_time']=df.apply(lambda row: row.termination_time - row.courier_started_order_local if row.termination_time - row.courier_started_order_local >= 0 else row.courier_started_order_local - row.termination_time + 86400 , axis=1)
df['Courier_capacity']=df.apply(lambda x: 3600.0/x.Total_time, axis=1)
```

Now let's plot the first two, which you asked for:

```python
(df['Delivery_time']/60).plot.hist(bins=50, title="Visibility of Delivery Time (minutes)")
```

```python
(df['Waiting_time_at_pickup']/60).plot.hist(bins=50, title="Visibility of Waiting Time at Pickup (minutes)")
```

And now, the last thing to do, is compute the Courier Capacity, for which first we are going to plot the distribution of the Total Time of an order (we excluded few results which where wrong on the data by setting the x_range):

```python
(df['Total_time']/60).plot.hist(bins=50, title="Visibility of Total Time (minutes)", range=[0,90])
```

The first thing that come to my head is to relate the reasignments with the Total Real Distance, because the couriers won't want to travel so much.

Or with the type of transport they are using, bicycle transporters will likely reject more orders in unevenness terrains than motorbikes, maybe.

To check this I will transform the Vertical and Transport into numericla data, so we can analyze it with a correlation matrix, first let's check the possible values of these two columns:

```python
print(df['vertical'].unique())
print(df['transport'].unique())
```

```python
df['vertical'] = df['vertical'].apply(lambda x: 0 if x=='WALL - Partner' else (1 if x== 'WALL - nonPartner' else (2 if x== 'QUIERO' else 3)))
df['transport'] = df['transport'].apply(lambda x: 0 if x=='BICYCLE' else (1 if x== 'MOTORBIKE' else 2))
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

corrMatrix = df.corr()

f =plt.figure(figsize=(19, 15))
sn.heatmap(corrMatrix, annot=True)
plt.title('Correlation Matrix', fontsize=25);
plt.show()
```

We see there are correlation with the vertical and the transport, as we hipotesized, but they are very small, or the order of 0.035.

We also see that there is a small correlation with the Total distance, and all the Order Times, of the order of 0.075-0.085 (we see that zone a bit brighter than the rest).

But the most correlated metric is Store Adress Id, with a correlation of 0.11, so we see that couriers tend to reject orders from given concrete places (maybe they don't treat them well, or are just too crowded, or don't have easy acces, etc...).
