# Central limit theorem

We generate samples of _Nsample_ values and calculate the mean. We collect the means for _N_ such samples and look at the distribution of these means.


https://en.wikipedia.org/wiki/Central_limit_theorem

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import statsmodels.distributions.empirical_distribution as stm

# Sampling parameters
Nsample = 500

# Distribution
dist = scp.beta(2,2)
# dist= scp.chi2(5)

# Make many samples and take the mean of each one
N = 1000
means= np.zeros(N)
for i in range(N):
    means[i] = np.mean(dist.rvs(size=Nsample)) 
    
# Plot the histogram, CDF
fig = plt.figure(frameon=False)
fig.set_size_inches(15,15)

p1= plt.subplot(2,2,1)
p= plt.hist(means,20)

p1= plt.subplot(2,2,2)
ecdf = stm.ECDF(means)
plt.plot(ecdf.x,ecdf.y,label = 'CDF empirical')

# QQ plot of the sample of means
# p1= plt.subplot(2,2,3)
# sm.qqplot(means, fit=True, line='45',plot=fig) # By default the comparison distribution is normal
```

## QQ plot to check normality of the distribution of means

```python
# QQ plot of the sample of means
sm.qqplot(means, fit=True, line='45') # By default the comparison distribution is normal
plt.show()
```

## Galton board

We simulate a Galton board by taking _Nrows_ random [0,1] values and summing them to represent the left-right bounces in each row.

https://www.youtube.com/watch?v=6YDHBFVIvIs

![](GaltonBoard.gif)

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
import statsmodels.distributions.empirical_distribution as stm

# Number of rows in the board
Nrows = 10

# Binomial distribution: 1 throw, 50% probability
dist = scp.binom(1,0.5)

# Simulate many dropping balls
N = 1000
sums= np.zeros(N)
for i in range(N):
    jumps = dist.rvs(Nrows) # this is the N jumps left-right of a ball
                            # looks like [0 0 1 0 0 0 1 1 0 0]
    jumps = np.where(jumps==0, -1, jumps) # Replace 0s by -1s
#     print(jumps, np.sum(jumps))
    sums[i] = np.sum(jumps) # Sum them = final position on the board
    

# Count instances of each value and do a barplot
values, counts = np.unique(sums, return_counts=True) 

fig = plt.figure(frameon=False)
fig.set_size_inches(15,15)
plt.bar(values,counts)
plt.xlabel("Sum")
plt.ylabel("N")


# Add a normal distribution as a reference
mu = np.mean(sums)
sig = np.sqrt(np.var(sums))
Cn =  np.max(counts) / scp.norm.pdf(0.,mu,sig)

p = plt.plot(values, Cn*scp.norm.pdf(values,mu,sig),'r-', lw=2, label='norm')
```

## Checking the theorem using joint distributions

We can also get an intuitive check of the central limit theorem from first principles, deriving the PDF of the sum of independent variables.

__Theorem:__ Let X and Y be  two  independent  random  variables  with PDfs $f_X(x)$ and $f_Y(y)$ defined for all x.  Then the sum Z=X+Y is a random variable with density function $f_Z(z)$, where $f_Z$ is the convolution of $f_X$ and $f_Y$:

$f_Z(z) = \int_{-\infty}^\infty f_X(z-y) f_Y(y) dy$

(see page 291 of https://www.dartmouth.edu/~chance/teaching_aids/books_articles/probability_book/Chapter7.pdf )

We can then calculate the PDF of the sum of two or more random variables for different cases by using the Python convolution tools.

(Examples by Adam Teixidó Bofill, master AFPC 2020-2021)

### Auxiliary functions

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np


def sum6(pdf, x): 
    """Returns the sum of up to 6 identical independent random variables, with PDF 
       the probability density evaluated at each x. x must be equispaced 
       and max(x) = -min(x)"""
    binSize = x[1]-x[0]
    binSizeAprox = 1/np.sum(pdf)
    C1 = pdf*binSize #Approximate probability to be in a bin (x[i],x[i+1])
    error = 1-binSize/binSizeAprox
    print("Initial error magnitude due to the tails and discretization:", 100*error, "%")
    
    span = x[-1]-x[0]
    bins = len(x)
    
    C2 = np.convolve(C1,C1,mode = "full")
    x2 = np.append(x[:-1], x+span)-span/2 #We need to add more x axis bins, approximately the same quantity on both sides.
    
    C3 = np.convolve(C2,C1,mode = "full")
    x3 = np.append(x2[:bins-1], x2+span)-span/2
    
    C4 = np.convolve(C3,C1,mode = "full")
    x4 = np.append(x3[:bins-1], x3+span)-span/2
    
    C5 = np.convolve(C4,C1,mode = "full")
    x5 = np.append(x4[:bins-1], x4+span)-span/2
    
    C6 = np.convolve(C5,C1,mode = "full")
    x6 = np.append(x5[:bins-1], x5+span)-span/2
    
    #print(len(C2),len(C3),len(C4),len(C5),len(C6))
    
    print("Error magnitude at the 6th convolution:", 100*(1-np.sum(C6)), "%")
    
    #Rescale the convolution results to get back pdf
    return [[x,C1/binSize], [x2,C2/binSize], [x3,C3/binSize], [x4,C4/binSize], [x5,C5/binSize],[x6,C6/binSize]] 
    
    
def weightedSum(Sum): 
    """Weight the sums of random variables with sqrt(n) to obtain the normalized variable Yn of the central limit theorem"""
    Yn = [[0,0] for i in range(len(Sum))]
    for i in range(len(Sum)):
        Yn[i][0] = Sum[i][0]/np.sqrt(i+1)
        Yn[i][1] = Sum[i][1]*np.sqrt(i+1)
    return Yn

    
def plots6(Data, subtitle, title):
    """Make six PDF plots with the x, y data inside each of the six entries of Data"""
    fig = plt.figure(frameon=False)
    fig.set_size_inches(15,7.5)
    plt.suptitle(title, size = 16)
    
    p1= plt.subplot(2,3,1)
    plt.plot(Data[0][0], Data[0][1],'r-', lw=2, label='norm')
    plt.title(subtitle[0])
    plt.ylabel('PDF')
    plt.xlabel('x')

    p1= plt.subplot(2,3,2)
    plt.plot(Data[1][0], Data[1][1],'r-', lw=2, label='norm')
    plt.title(subtitle[1])
    plt.ylabel('PDF')
    plt.xlabel('x')

    p1= plt.subplot(2,3,3)
    plt.plot(Data[2][0], Data[2][1],'r-', lw=2, label='norm')
    plt.title(subtitle[2])
    plt.ylabel('PDF')
    plt.xlabel('x')

    p1= plt.subplot(2,3,4)
    plt.plot(Data[3][0], Data[3][1],'r-', lw=2, label='norm')
    plt.title(subtitle[3])
    plt.ylabel('PDF')
    plt.xlabel('x')

    p1= plt.subplot(2,3,5)
    plt.plot(Data[4][0], Data[4][1],'r-', lw=2, label='norm')
    plt.title(subtitle[4])
    plt.ylabel('PDF')
    plt.xlabel('x')

    p1= plt.subplot(2,3,6)
    plt.plot(Data[5][0], Data[5][1],'r-', lw=2, label='norm')
    plt.title(subtitle[5])
    plt.ylabel('PDF')
    plt.xlabel('x')

    plt.tight_layout()
    plt.subplots_adjust(top=0.90)
    plt.show()
    
    
def plotCompare(Data, title): 
    """Plot the 6 PDF inside Data toghether with a Gaussian N(0,1)"""
    fig = plt.figure(frameon=False)
    fig.set_size_inches(15,7.5)
    plt.title(title, size = 16)
   
    plt.plot(Data[0][0], Data[0][1],'r-', lw=2, label='Y1')
    plt.plot(Data[1][0], Data[1][1],'r-.', lw=2, label='Y2')
    plt.plot(Data[2][0], Data[2][1],'r:', lw=2, label='Y3')
    plt.plot(Data[3][0], Data[3][1],'r--', dashes=(2,2), lw=2, label='Y4')
    plt.plot(Data[4][0], Data[4][1],'r--', dashes=(5,5), lw=1, label='Y5')
    plt.plot(Data[5][0], Data[5][1],'r--', dashes=(8,3), lw=1, label='Y6')
    plt.plot(Data[5][0], scp.norm.pdf(Data[5][0],0,1),'k-', lw=2, label='N(0,1)')
    plt.ylabel('PDF')
    plt.xlabel('x')
    
    if(Data[5][0][-1]-Data[5][0][0] > 20): #Put a maximum limit on the values of the x axis.
        plt.xlim((-10,10))
    
    plt.legend()
    plt.show()
```

### Uniform distribution

```python
#Define x axis
N = 500
min = -3.
max = -min #The sum6 function which is used afterwards only works for symmetric x intervals 
x = np.linspace(min, max, N) # N points from min to max

#Define the distributions
Xi = scp.uniform.pdf(x, loc = -np.sqrt(12)/2, scale = np.sqrt(12))
Sum = sum6(Xi, x) #PDFs of random variables X1, X1+X2, ..., X1+...+X6.
Yn = weightedSum(Sum) #PDFs of Y1 to Y6. Central limit theorem: Yn random variable which tends to N(0,1). 

# Plot PDFs 
plots6(Sum, ['$X_1$','$X_1+X_2$','$X_1+X_2+X_3$','$X_1+X_2+X_3+X_4$','$X_1+X_2+X_3+X_4+X_5$','$X_1+X_2+X_3+X_4+X_6$'], "$X_i$ = Uniform random variable (Normalized)")
plotCompare(Yn,"Limit of $Y_n = (X_1+...+X_n)/\sqrt{n}$,  $X_i$ = Uniform random variable (Normalized)")


#The Yn tend to the normalized Gaussian, N(0,1).
```

### Beta distribution

```python
#Define x axis
N = 1000
min = -2.
max = -min
x = np.linspace(min, max, N) # N points from min to max

#Define the distributions
a, b = 0.85,0.75
mu = a/(a+b) 
sigma = np.sqrt(a*b/((a+b)**2*(a+b+1)))
Xi = scp.beta.pdf(x*sigma+mu,a, b)*sigma #For simplicity, we take the distribution to have mean = 0 and variance = 1.
Sum = sum6(Xi, x) #PDFs of random variables X1, X1+X2, ..., X1+...+X6.
Yn = weightedSum(Sum) #PDFs of Y1 to Y6. Central limit theorem: Yn random variable which tends to N(0,1). 

# Plot PDFs 
plots6(Sum, ['$X_1$','$X_1+X_2$','$X_1+X_2+X_3$','$X_1+X_2+X_3+X_4$','$X_1+X_2+X_3+X_4+X_5$','$X_1+X_2+X_3+X_4+X_6$'], "$X_i$ = Beta random variable (normalized)")
plotCompare(Yn,"Limit of $Y_n = (X_1+...+X_n)/\sqrt{n}$,  $X_i$ = Beta random variable (normalized)")


#The Yn tend to the normalized Gaussian, N(0,1).
```

### $\chi^2$ distribution

```python
#Define x axis
N = 1000
min = -15.
max = -min
x = np.linspace(min, max,N) # N points from min to max

#Define the distributions
k = 3
mu = k
sigma = np.sqrt(2*k)
Xi = scp.chi2.pdf(x*sigma+mu,k)*sigma #For simplicity, we take the distribution to have mean = 0 and variance = 1.
Sum = sum6(Xi, x) #PDFs of random variables X1, X1+X2, ..., X1+...+X6.
Yn = weightedSum(Sum) #PDFs of Y1 to Y6. Central limit theorem: Yn random variable which tends to N(0,1). 

# Plot PDFs 
plots6(Sum, ['$X_1$','$X_1+X_2$','$X_1+X_2+X_3$','$X_1+X_2+X_3+X_4$','$X_1+X_2+X_3+X_4+X_5$','$X_1+X_2+X_3+X_4+X_6$'], "$X_i = \chi^2$ random variable (normalized)")
plotCompare(Yn,"Limit of $Y_n = (X_1+...+X_n)/\sqrt{n}$,  $X_i$ = $\chi^2$ random variable (normalized)")


#We get a slow convergence to the Gaussian N(0,1)
```
