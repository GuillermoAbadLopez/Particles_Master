# Law of large numbers

https://en.wikipedia.org/wiki/Law_of_large_numbers

```python
%matplotlib inline
import scipy.stats as scp
import matplotlib.pyplot as plt
import numpy as np

N = 500 # Number of points for each batch generation

mean = []
median = []
variance = []
all_points = []

# Accumulate points and re-calculate statistics at each step
for i in range(600):
    all_points.extend(scp.norm.rvs(0.,1.,N))
#     all_points.extend(scp.chi2.rvs(5.,size=N))
#    all_points.extend(scp.beta.rvs(2.,2.,size=N))
    
    mean.append( np.mean(all_points) )
    median.append( np.median(all_points) )
    variance.append( np.var(all_points) )

fig = plt.figure(frameon=False)
fig.set_size_inches(15,15)

# Plot series
p1= plt.subplot(2,2,1)
plt.plot(mean, 'r.',)
plt.title('Mean')
plt.xlabel('N')

p1= plt.subplot(2,2,2)
plt.plot(median, 'r.',)
plt.title('Median')
plt.xlabel('N')

p1= plt.subplot(2,2,3)
plt.plot(variance, 'r.',)
plt.title('Variance')
plt.xlabel('N')

# Plot histograms
p1= plt.subplot(2,2,4)
plt.hist(all_points,100)
plt.title('Full histogram')

# p1= plt.subplot(2,2,4)
# plt.hist(all_points,100,range=(-7,7))
# plt.title('Clipped histogram')

plt.show()
```
